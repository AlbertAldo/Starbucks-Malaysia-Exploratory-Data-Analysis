```python
barista = ['Barista Coffee and Colddrinks', 'Cashier']
gaji = [4.3,3.7]
totalstaff = [6,5]
bayaran = pd.DataFrame(gaji,barista, columns=['Gaji per Jam'])
bayaran['Gaji per Shift'] = bayaran['Gaji per Jam'] * 6
bayaran['Total Staff'] = totalstaff
bayaran['Total Gaji per Hari'] = bayaran['Gaji per Shift'] * bayaran['Total Staff']
bayaran['Total Gaji per Bulan'] = bayaran['Total Gaji per Hari']*30
```

```python
shiftmalam = [1,1]
```

```python
bayaran['Staff Shift 1'] = shiftmalam
bayaran['Penghematan Shift 1 Selama 30 hari'] = bayaran['Staff Shift 1'] * bayaran['Gaji per Shift'] * -1 * 30
bayaran.reset_index(inplace=True)
bayaran
```

```python
barista = ['Barista Coffee and Colddrinks', 'Cashier']
gaji = [4.3,3.7]
totalstaff = [5,4]
bayaran2 = pd.DataFrame(gaji,barista, columns=['Gaji per Jam'])
bayaran2['Gaji per Shift'] = bayaran2['Gaji per Jam'] * 6
bayaran2['Total Staff'] = totalstaff
bayaran2['Total Gaji per Hari'] = bayaran2['Gaji per Shift'] * bayaran2['Total Staff']
bayaran2['Total Gaji per Bulan'] = bayaran2['Total Gaji per Hari']*30
bayaran2.reset_index(inplace=True)
bayaran
```

```python
plt.figure(figsize=(10,8))
plt.bar(bayaran['index'], bayaran['Total Gaji per Bulan'], 
        color='r', 
        edgecolor ='black', 
        linewidth = 3, width = 0.3)

for i in bayaran['Total Gaji per Bulan']:
    X = bayaran[bayaran['Total Gaji per Bulan'] == i]['index']
    plt.text(X,i+30,round(i,0), fontdict={'color' : 'black', 'size' : 12})

plt.bar(bayaran2['index'], bayaran2['Total Gaji per Bulan'], 
        color='b', 
        edgecolor ='black', 
        linewidth = 3, width = 0.3,align='edge')

for i in bayaran2['Total Gaji per Bulan']:
    X = bayaran2[bayaran2['Total Gaji per Bulan'] == i]['index']
    plt.text(X,i+30,round(i,0), fontdict={'color' : 'black', 'size' : 12})

plt.legend(['Keadaan Normal', 'Setelah Shift 1 ditiadakan'], loc = 0)
plt.title('Pengehematan Setelah Shift 1 dihilangkan', fontdict={'size':'12'})
plt.xlabel('Karyawan', fontdict={'size':'12'})
plt.ylabel('Gaji (RM)', fontdict={'size':'12'})
plt.savefig('Pengehematan_Setelah_Shift_1_dihilangkan.png')
plt.show()
```
